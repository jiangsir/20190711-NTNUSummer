### 我們發現，以上的乘積表沒有座號，難以排序，因此成績不容易登記。

### 我們希望資料當中的"學號"能夠自動對應出班級座號

```python
# 首先，先看看全校名條的格式
import pandas as pd
df_source = pd.read_excel(basepath+'107-2--全校名條.xlsx', sheet_name='高中名條')

df_source.head(10)
```

```python
import pandas as pd

basepath = '.\\datas\\'
csvfiles = [
'107-3.2-高二和資訊科技(1).csv',
'107-3.2-高二和資訊科技(2).csv',
'107-3.2-高二和資訊科技(3).csv',
'107-3.2-高二和資訊科技(4).csv'
]

df_source = pd.read_excel(basepath+'107-2--全校名條.xlsx', sheet_name='高中名條')

for csvfile in csvfiles:
    # 先用 big5 嘗試開啟，若發生例外，再改用 UTF-8 開啟，這樣就可以將兩種編碼的檔案都順利打開
    try:
        df_scoreboard = pd.read_csv(basepath + csvfile, engine='python', encoding='big5')
    except Exception as e:
        df_scoreboard = pd.read_csv(basepath + csvfile, engine='python', encoding='utf-8')
    df_scoreboard['班級座號'] = pd.Series()
    for index, row in df_scoreboard.iterrows():
        try:
            studentid = row.account.split('@')[0]
            loc = df_source.loc[df_source['學號'] == int(studentid)]
            df_scoreboard['班級座號'][index] = str(loc['班級'].values[0]) + str(f'{loc["座號"].values[0]:02d}' )
        except Exception as e:
            # 遇到無法對應的資料，就直接跳到下一個。
            print(e)
            continue


    df_scoreboard.to_csv(basepath+'new_'+csvfile,index=False,sep=',',encoding='utf-8-sig')
    print(df_scoreboard)
```
