## 跨頁連續讀取

````python
import requests
from bs4 import BeautifulSoup

host = 'https://www.ptt.cc'
url = '/bbs/Beauty/index.html'
pages = 10

def readPage(url):
    print('開始讀取頁面:', url)
    result = requests.get(url)
    soup = BeautifulSoup(result.text, 'lxml')
    selector = "div.title a"
    tags = soup.select(selector) # 取得一組 bs4.element.Tag 的 list
    for tag in tags:
        print(host+tag['href'], tag.text)


readPage(host+url)

for i in range(pages):
    result = requests.get(host + url)
    soup = BeautifulSoup(result.text, 'lxml')

    # 面對多個 class name 的兩種寫法
    selector = "div[class='btn-group btn-group-paging'] a"
    #selector = "div.btn-group.btn-group-paging a"

    tags = soup.select(selector) # 取得一組 bs4.element.Tag 的 list
        
    url = tags[1]['href']
    print(url)
    readPage(host+url)

    
````

## 抓取頁面內的圖片

````python
import requests
from bs4 import BeautifulSoup
import re
from IPython.display import Image
from IPython.core.display import HTML
from IPython.core.display import Image, display

host = 'https://www.ptt.cc'
url = '/bbs/Beauty/index.html'
pages = 1
regex_img = re.compile('http[s]?://i.imgur.com/\w+\.(?:jpg|png|gif)')
regex_img2 = re.compile('http[s]?://imgur.com/\w+')

def readPage(url):
    print('開始讀取頁面:', url)
    result = requests.get(url)
    soup = BeautifulSoup(result.text, 'lxml')
    selector = "div.title a"
    tags = soup.select(selector) # 取得一組 bs4.element.Tag 的 list
    for tag in tags:
        print(host+tag['href'], tag.text)
        readImage(host+tag['href'], tag.text)


def readImage(pageUrl, title):
    result = requests.get(pageUrl)
    images = regex_img.findall(result.text)
    image2s = regex_img2.findall(result.text)
    image3s = []
    for image2 in image2s:
        image3s.append(image2+".jpg")
    images = set(images).union(set(image3s))
    print(*images)
    for image in images:
        print(title, image)
        display(Image(url= image))

    
readPage(host+url)

for i in range(pages):
    result = requests.get(host + url)
    soup = BeautifulSoup(result.text, 'lxml')

    # 面對多個 class name 的兩種寫法
    selector = "div[class='btn-group btn-group-paging'] a"
    #selector = "div.btn-group.btn-group-paging a"

    tags = soup.select(selector) # 取得一組 bs4.element.Tag 的 list
        
    url = tags[1]['href']
    print(url)
    readPage(host+url)
````

## 下載頁面內的圖片

增加以下 download 函數，給定 url 及下載後所需要的檔名，即可下載。

```python
def download(url, filename):
    r = requests.get(url)
    filename = os.path.join(downloadpath, filename)
    with open(filename,'wb') as f:     #二進位制寫入
        f.write(r.content)
    print(url, '-->', filename)

```

呼叫方式:
用 rindex 從尾端搜尋最後一個 / 以便取得檔名，捨去網址路徑。

```python
    download(image, title+image[image.rindex('/')+1:])

```

````python
import requests, os
from bs4 import BeautifulSoup
import re

host = 'https://www.ptt.cc'
url = '/bbs/Beauty/index.html'
pages = 1
regex_img = re.compile('http[s]?://i.imgur.com/\w+\.(?:jpg|png|gif)')
regex_img2 = re.compile('http[s]?://imgur.com/\w+')

downloadpath = '.\\downloads\\'

def download(url, filename):
    r = requests.get(url)
    filename = os.path.join(downloadpath, filename)
    with open(filename,'wb') as f:     #二進位制寫入
        f.write(r.content)
    print(url, '-->', filename)


def readPage(url):
    print('開始讀取頁面:', url)
    try:
        result = requests.get(url)
        soup = BeautifulSoup(result.text, 'lxml')
        selector = "div.title a"
        tags = soup.select(selector) # 取得一組 bs4.element.Tag 的 list
        for tag in tags:
            print('文章: ', tag.text, host+tag['href'])
            readImage(host+tag['href'], tag.text)
    except Exception as e:
        pass

def readImage(pageUrl, title):
    result = requests.get(pageUrl)
    images = regex_img.findall(result.text)
    image2s = regex_img2.findall(result.text)
    image3s = []
    for image2 in image2s:
        image3s.append(image2+".jpg")
    images = set(images).union(set(image3s))
    #print(*images)
    for image in images:
        #display(Image(url= image))
        download(image, title+image[image.rindex('/')+1:])

    
readPage(host+url)

for i in range(pages):
    result = requests.get(host + url)
    soup = BeautifulSoup(result.text, 'lxml')

    # 面對多個 class name 的兩種寫法
    selector = "div[class='btn-group btn-group-paging'] a"
    #selector = "div.btn-group.btn-group-paging a"

    tags = soup.select(selector) # 取得一組 bs4.element.Tag 的 list
        
    url = tags[1]['href']
    print(url)
    readPage(host+url)
````
