# 實作範例 4 - 非監督式學習 Unsupervised Learning

上一個章節的實作範例中，我們利用了 Scikit-learn 中的 `KNeighborsClassifier` 與 `DecisionTreeClassifier` 模組，將資料使用 *kNN* 與*決策樹*兩種監督式學習的演算法進行了分類 (classification)。本實作範例中，我們將繼續學習使用 Scikit-learn 套件，對資料進行「**分群 (clustering)**」。「分群」與「分類」相比，不同的是訓練資料中並**沒有**已經標示好的類別，也就是沒有可以參考的資訊，需要直接利用資料本身的特徵分類，而且通常需自行指定群的數量。

本實作範例中將會繼續介紹使用 scikit-learn 來實作兩個常見的非監督式演算法：**k-平均演算法 (k-means algorithm)**與**階層式分群法 (hierarchical clustering)**。

首先我們載入必要的 Python 套件。與之前相同，我們除了使用 scikit-learn (sklearn) 之外，也利用 matplotlib 來畫圖：

```python
import matplotlib.pyplot as plt # 載入繪圖套件
from sklearn import datasets # 載入 scikit-learn 套件中的資料集模組，本章範例將用來產生資料集
from sklearn import cluster # 載入 scikit-learn 套件中的分群演算法模組
```

## k-平均演算法 (k-means algorithm)

### 產生資料

首先我們利用 scikit-learn 中的 `datasets.make_blobs` 來產生一組資料，然後透過 matplotlib 畫出來看看：

```python
# 隨機產生有 4 個群的資料
points, label = datasets.make_blobs(n_samples=1500, centers=4, cluster_std=1.6, random_state=170)

# 印出資料並且依據類別給予顏色
plt.figure(figsize=(6, 6)) # 設定圖片的大小，這裡的 6, 6 是指圖片的長/寬各 6 英吋
plt.scatter(points[:, 0], points[:, 1], c=label[:])
plt.show()
```

我們可以看到，`make_blobs()` 可以*隨機*產生一群一群圓形的資料，而這個函式除了產生資料外，也會順便幫我們標記好每個點的類別，因此我們需要用兩個變數 (上面的 `points` 與 `labels`) 去接 `make_blobs()` 回傳的資料。

`make_blobs()` 有幾個主要參數可以調整資料的樣子：
 * `n_samples`: 資料點的數量
 * `n_features`: 特徵的數量，預設是 2，也就是每個資料點有兩個值
 * `centers`: 群的數量
 * `cluster_std`: 群中資料點離群中央的距離標準差，這個值愈大資料愈分散
 * `random_state`: 如果這個值在每次呼叫 `make_blobs()` 都一樣的話，會產生一樣的結果。專業術語上可視為[偽隨機數生成器](https://zh.wikipedia.org/zh-tw/偽隨機數生成器)的*隨機種子*

根據參數，上面的資料是一個有 1500 個點，並且每個點有 2 個特徵的資料，所以 `points` 變數就是一個 $1500 \times 2$ 的 `ndarray`；而每個點都對應到一個類別，故 `labels` 會是一個 $1500 \times 1$ 的 `ndarray`。

關於 `make_blobs()` 的詳細用法，請參考[官方文件](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html)。

然後在 `pyplot.scatter()` 中，我們可以將資料的類別丟到 `c` 這個參數中，讓 matplotlib 自動幫我們將不同的群標上不同的顏色，方便我們觀察。

### 利用 scikit-learn分群

接著我們將使用 scikit-learn 提供的 `cluster.KMeans` 模組將上面所產生的資料利用 k-means 演算法分群。

`cluster.KMeans` 與其他 scikit-learn 提供的內建演算法相同，也是先建立 `cluster.KMeans` 物件後使用 `fit()` &rarr; `predict()` 這個流程：

```python
# 初始化 k-means 分群模組
k_mean = cluster.KMeans(n_clusters=4)

# 開始分群
y_pred_kmean = k_mean.fit_predict(points)

# 畫出分群結果
plt.figure(figsize=(6, 6))
plt.scatter(points[:, 0], points[:, 1], c=y_pred_kmean[:])
plt.show()
```

首先在 `cluster.KMeans` 的初始化中有幾個重要的參數可以設定：
* `n_clusters`: 指定要分成幾群
* `n_init`: 要跑幾次 k-means 分群，預設是 10 次。由於 k-means 中的群中央在一開始都是隨機給定的，為了避免一開始選到不好的中央點，我們可以使用不同的隨機中央點跑很多次，然後挑一次最好的當做結果
* `max_iter`: 每一次計算群中心點時的最大迭代次數，預設是 300 次。在某些情況中，群中心點可能不會收斂，會交互在某幾個點中不斷循環。透過這個參數，可以在發生這種情況時停止，避免無限迴圈

然後我們直接使用 `fit_predict()` 將資料點分群。

同學們可以試著調整資料與演算法的參數，觀察看看分群前後的變化喔！

關於 `cluster.KMeans` 的詳細用法，有興趣的讀者請參閱[官方文件](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html)。

## 階層式分群法 (hierarchical clustering)

階層是分群通常分為兩種作法：**聚合法 (agglomerative, bottom-up, 小群到大群)** 與**分裂法 (divisive, top-down, 大群到小群)** 兩種，本範例中將介紹聚合型的方法。

我們沿用之前所產生的資料，然後使用 `cluster.AgglomerativeClustering` 進行分群：

```python
# 初始化 AgglomerativeClustering 分群模組
hierarchical = cluster.AgglomerativeClustering(n_clusters=4)

# 開始分群
y_pred_h = hierarchical.fit_predict(points)

# 畫出分群結果
plt.figure(figsize=(6, 6))
plt.scatter(points[:, 0], points[:, 1], c=y_pred_h[:])
plt.show()
```

基本上 `cluster.AgglomerativeClustering` 的*使用方式*與 `cluster.KMeans` 相同；但是因演算法性質不同，可調整的參數不太一樣。以下列出幾個重要的參數：
* `n_clusters`: 與 `cluster.KMeans` 相同，指定要分成幾群
* `linkage`: 各群間距離的計算方式 ([linkage criteria](https://en.wikipedia.org/wiki/Hierarchical_clustering#Linkage_criteria))，有以下幾個值可以使用：
  - `ward` (預設值): 利用 [Ward's method](https://en.wikipedia.org/wiki/Ward%27s_method) 執行群結合，其意義是在挑選兩群結合時，能夠讓結合後的群中之變異數 (各點間之距離和) 最小
  - `average`: 以兩群中各點間之**平均**距離作為兩群之距離
  - `complete`: 以兩群中各點間之**最長**距離作為兩群之距離
  - `single`: 以兩群中各點間之**最短**距離作為兩群之距離

同學們可以試著調整資料與演算法的參數，觀察分群前後的變化喔。

關於 `cluster.AgglomerativeClustering` 的詳細用法，有興趣的讀者請參閱[官方文件](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.AgglomerativeClustering.html)。

關於 scikit-learn 中關於分群主題的介紹，請參考官方的[使用指南](https://scikit-learn.org/stable/modules/clustering.html)。

## 同場加映：不一樣的資料

k-means 與階層式分群法都是基於**距離**的演算法，這類型的分群演算法適合如上面群間距離單純的資料；如果遇到不能單純使用距離來計算群間的*相似度*時，會發生什麼事呢？

我們來看看下面的資料：

```python
# 產生半月形資料
moon_data, moon_data_labels = datasets.make_moons(n_samples=1500, noise=0.05)

# 畫出來看看
plt.scatter(moon_data[:, 0], moon_data[:, 1], c=moon_data_labels)
plt.show()
```

這類型的資料較難單純利用距離來執行分群，我們來看看若使用上面所介紹的演算法會發生什麼事：

```python
# 利用 k-means 分群
km_moon = cluster.KMeans(n_clusters=2)
moon_data_labels_km = km_moon.fit_predict(moon_data)
plt.scatter(moon_data[:, 0], moon_data[:, 1], c=moon_data_labels_km[:])
plt.show()
```

```python
# 利用階層式分群法分群
agg_moon = cluster.AgglomerativeClustering(n_clusters=2)
moon_data_labels_agg = agg_moon.fit_predict(moon_data)
plt.scatter(moon_data[:, 0], moon_data[:, 1], c=moon_data_labels_agg[:])
plt.show()
```

同學們有沒有發現什麼呢？如果我們指定利用階層式分群法中的 linkage 為 **single** ([single-linkage clustering](https://en.wikipedia.org/wiki/Single-linkage_clustering)) 的話呢？

```python
# 利用階層式分群法 (single-linkage clustering) 分群
aggsl_moon = cluster.AgglomerativeClustering(n_clusters=2, linkage='single')
moon_data_labels_aggsl = aggsl_moon.fit_predict(moon_data)
plt.scatter(moon_data[:, 0], moon_data[:, 1], c=moon_data_labels_aggsl[:])
plt.show()
```

同學們可以想想，是什麼造成了這種差異呢？

讓我們再看看這種資料：

```python
# 產生環狀資料
circle_data, circle_data_labels = datasets.make_circles(n_samples=1500, noise=0.01)

# 畫出來看看
plt.scatter(circle_data[:, 0], circle_data[:, 1], c=circle_data_labels)
plt.show()
```

同學們也可以想想，為什麼會這樣呢？
